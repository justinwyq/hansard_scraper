```python
import json
import requests
import pandas as pd
import time

pd.options.display.max_columns = 999
```

```python
## Parliament website has no anti scraping protection
## Hence, post or get request will suffice for web scraping. Faster and less error prone than selenium
## all reports will be extracted in json format and stored in pandas dataframe (saved in pickle format)
```

```python
def get_index(parlno, start):

    # gets the indices of hansards for specified parliament number
    # tries 3 times in case of errors
    
    for i in range(3):
        try:
            headers = {
                'Content-type': 'application/json; charset=UTF-8'
                }
            payload = {
              "keyword": "undefined",
              "fromday": "01",
              "frommonth": "04",
              "fromyear": "2023",
              "today": "01",
              "tomonth": "04",
              "toyear": "2023",
              "dateRange": "* TO NOW",
              "reportContent": "with all the words",
              "parliamentNo": f"{parlno}",
              "selectedSort": "date_dt desc",
              "portfolio": [],
              "mpName": "",
              "rsSelected": "",
              "lang": "",
              "startIndex": f"{start}",
              "endIndex": "",
              "titleChecked": "false",
              "footNoteChecked": "false",
              "ministrySelected": []
            }

            url = "https://sprs.parl.gov.sg/search/searchResult"
            res = requests.post(url, data=json.dumps(payload), headers=headers)

            return res.json()
        except:
            time.sleep(5)
            continue
        else:
            print(f"no results found at index {start} at parliament number {parlno}")
            print("status code: " + res.status_code)
```

```python
# total reports in the parliament sitting is unknown, so set it to 20 first
total_reports = 20
i = 0
df = pd.DataFrame()
while i < total_reports:
    js = get_index(0, i)
    total_reports = int(js[0]['maxResult'])
    df = df.append(pd.json_normalize(js))
    print(f"now scraping batch {i}")
    i+=20
```

```python
df = df.set_index('sno')
df.to_pickle('parl0.pkl')
```

```python
# now do the other parl sessions:

for j in range(1,15):
    total_reports = 20
    i = 0
    df = pd.DataFrame()
    while i < total_reports:
        js = get_index(j, i)
        total_reports = int(js[0]['maxResult'])
        df = df.append(pd.json_normalize(js))
        print(f"now scraping batch {i} for parl number {j}")
        i+=20
    df = df.set_index('sno')
    df.to_pickle(f'parl{j}.pkl')
```

```python
def get_report(reportVersion, htmlFileName, reportId):

    # get full details of each report. 
    # url format depends on reportVersion (the version changed from sprs2 to sprs3 some time in parliament #12)
    # tries 3 times in case of errors
    
    for i in range(3):
        try:
            if reportVersion == 'sprs2':
                url = f'https://sprs.parl.gov.sg/search/getHansardTopic/?id={htmlFileName}'
            elif reportVersion == 'sprs3':
                url = f'https://sprs.parl.gov.sg/search/getHansardTopic/?id={reportId}'
            else:
                print('fatal error, unknown report version')
                break
            res = requests.get(url)
            return res.json()
        
        except:
            time.sleep(5)
            continue
            
        else:
            print(f"no results found for report {htmlFileName}{reportId}")
            print("status code: " + res.status_code)
```

```python
# now loop through all the parliament indices and get all report details

for i in range(15):
    #read index file
    df = pd.read_pickle(f'parl{i}.pkl')
    
    #initialise empty df
    df_report = pd.DataFrame()
    
    # iterate through all rows and get all report details
    # use numpy to iterate as it's much faster than iterrows
    for row in df[['reportVersion', 'htmlFileName', 'reportId']].values:
        print(f'now scraping parl no {i}, report {row[1]}{row[2]}')
        js_report = get_report(row[0], row[1], row[2])
        
        # try-except needed here because there are some reports that have broken link from Parl website
        # for example, report https://sprs.parl.gov.sg/search/#/topic?reportid=000_3_20040105_S0001
        # if broken link, the json will be empty and pd.json_normalize will throw error
        try:
            df_report = df_report.append(pd.json_normalize(js_report))
        except:
            # save errored rows into a file for further troubleshooting, use append mode
            df[(df.htmlFileName==row[1])|(df.reportId==row[2])].to_csv('scrape_error_reports.csv', mode='a')
            continue
        
        
    #save report details
    df_report.to_pickle(f'hansard_details_parl{i}.pkl')
```
