```python
import json
import requests
import pandas as pd
```

```python
# session = Session()

headers = {
    'Content-type': 'application/json; charset=UTF-8'
    }
payload = {
  "keyword": "undefined",
  "fromday": "01",
  "frommonth": "04",
  "fromyear": "2023",
  "today": "01",
  "tomonth": "04",
  "toyear": "2023",
  "dateRange": "* TO NOW",
  "reportContent": "with all the words",
  "parliamentNo": "14",
  "selectedSort": "date_dt desc",
  "portfolio": [],
  "mpName": "",
  "rsSelected": "",
  "lang": "",
  "startIndex": "0",
  "endIndex": "",
  "titleChecked": "false",
  "footNoteChecked": "false",
  "ministrySelected": []
}

url = "https://sprs.parl.gov.sg/search/searchResult"
res = requests.post(url, data=json.dumps(payload), headers=headers)
print(res.content)
```
